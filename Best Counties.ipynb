```python
#Data from https://www.usnews.com/news/healthiest-communities/washington
cnty = pd.read_csv('HealthyCounties.csv')
cnty
```

```python
cnty.info()
```

```python
#Dropped NAN and Unnamed cells, change order of columns
cnty_df = cnty.iloc[:, [0,10,8,12,14,16,2,18,4,20]]
cnty_df = cnty_df.dropna(how='any')
cnty_df.head()
```

```python
#Create columns for weighted totals
cnty_df['Housing Weight'] = cnty_df['Housing']*14
cnty_df['Economy Weight'] = cnty_df['Economy']*14
cnty_df['Food Weight'] = cnty_df['Food & Nutrition']*13
cnty_df['Environment Weight'] = cnty_df['Environment']*13
cnty_df['Public Safety Weight'] = cnty_df['Public Safety']*12
cnty_df['Pop Health'] = cnty_df['Population Health']*10
cnty_df['Vitality Weight'] = cnty_df['Community Vitality']*10
cnty_df['Equity Weight'] = cnty_df['Equity']*7
cnty_df['Infrastructure Weight'] = cnty_df['Infrastructure']*7
cnty_df
```

```python
#Create column of weighted average of best county traits
cnty_df['Healthy Counties Weighted Average'] = (cnty_df.loc[:, 'Housing Weight':'Infrastructure Weight'].sum(axis=1))/100
cnty_df.sort_values(by=['Healthy Counties Weighted Average'], ascending=False)
```
