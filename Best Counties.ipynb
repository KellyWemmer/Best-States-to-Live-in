```python
#Create column for state abbreviation to differentiate between states with same county names
state = []
for value in FIPS["State Code (FIPS)"]: 
    if value == 8: 
        state.append("CO") 
    elif value == 16: 
        state.append("ID") 
    elif value == 30: 
        state.append("MT")
    elif value == 41: 
        state.append("OR")
    elif value == 53: 
        state.append("WA")
    elif value == 56: 
        state.append("WY")
    else: 
        state.append("Not Found") 
       
FIPS["State"] = state    
FIPS.info()
```
